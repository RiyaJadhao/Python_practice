```python
# Arithmetic : +, -, * , /,//,%,**
# Comparison : > < >= <= == !=
# Logical: and or not
# Assignmet: = += -= *= /=
##------------------------------------
a, b= 10 , 3
print(a+b, a-b, a/b, a//b, a*b, a**b, a%b)
print(a>b)
print(a<b)
print(a>b, b>a)
print(a>b or b<5)
print(a>b and b<5)
print(a>b and b>a)
print(a>b or b>a)
```

```python
# Conditional Statement 
num=int(input("Enter number:"))
if num > 0:
    print("Positive")
elif num==0:
    print("Zero")
else:
    print("Negative")
###---------------------------
#Ternary
print("Even" if num%2==0 else "Odd")
```

```python
# LOOPS
#FOR LOOP 
for i in range(1,11):
    print(i, end=" ")
print()
for i in range(5): # start -0 
    print(i)
```

```python
# While Loop
n=10   #initalization 
while n>0:   #cond
    print(n)
    n-=1     #decrement
```

```python
# Loop Control
for i in range(1,10):
    if i == 5: continue
    if i == 8: break
    print(i)
```
