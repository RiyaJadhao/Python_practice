```python
# Write a Python program to check if each number is prime in a given list of numbers.
# Return True if all numbers are prime otherwise False.
def prime(n):
    if n < 2:
        return False
    for j in range(2, int(n**0.5) + 1):
        if n % j == 0:
            return False
    return True

def all_prime(lt):
    for i in lt:
        if not prime(i):   # checks each number using prime()
            return False   # if one is not prime → return False
    return True            # if loop completes → all are prime

l1 = [0, 3, 4, 7, 9]
l2 = [3, 5, 7, 13]
l3 = [1, 5, 3]

print(all_prime(l1))  # False
print(all_prime(l2))  # True
print(all_prime(l3))  # False
```

```python
#Write a Python program to flatten a simple list.
l = [10, 20, [3, 4, 5], 6, [11, 45], 8]
flat = []

for i in l:
    if type(i) == list:      # if element is a list
        for j in i:
            flat.append(j)
    else:                    # if element is not a list
        flat.append(i)

print(flat)
```

```python
l=[10,20,[3,4,5],6,[11,45],8]
flat=[]

for i in l:
    if isinstance(i,list):
        flat.extend(i)
    else:
        flat.append(i)
print(flat)
        
```
