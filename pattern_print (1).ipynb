# Pascalâ€™s Triangle

n=5
for i in range(n):
    print(" "*(n-i),end="")
    num=1
    for j in range(i+1):
        print(num,end=" ")
        num=num * (i-j) // (j+1)
    print()

# Floyd triangle

```python

n=5
num=1
for i in range(1,n+1):
    for j in range(1,i+1):
        print(num,end=" ")
        num += 1
    print()
```

```python
n = 5
num = 1
for i in range(1, n + 1):
    print(" " * (n - i), end="")
    for j in range(1, i + 1):
        print(f"{num} ", end="")
        num += 1
    print()
```
