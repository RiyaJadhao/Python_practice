```python
def rev(n):
    rev=0
    num=n
    while n>0:
        digit=n%10
        rev=rev *10+digit
        n=n//10
    print(rev)
n1=int(input("e:"))
rev(n1)
```

```python
def perfect(n):
    sum=0
    for i in range(1,n):
        if n%i==0:
            sum+=i
    if sum==n:
        print("perfect")
    else:
        print("not")
n1=int(input("e:"))
perfect(n1)
```

```python
def prime(n):
    if n > 1:
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                print("not")
                break
        else:
            print("prime")
    else:
        print("not")
n1=int(input("enter:"))
prime(n1)
```

```python
def palin(n):
    n=str(n)
    if n[::-1]==n:
        print("palindrome")
    else:
        print("n")
n1=int(input("enter:"))
palin(n1)
```

```python
#Count Even, Odd, Positive, Negative Digits
num = int(input("Enter number: "))
even = odd = 0
n = abs(num)
while n > 0:
    digit = n % 10 #159 -> 15
    if digit % 2 == 0:  
        even += 1
    else:
        odd += 1
    n //= 10  #159//10=15 

if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero")

print("Even digits:", even)
print("Odd digits:", odd)
```

```python
#Find GCD & LCM
#GCD (Greatest Common Divisor)
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a
def lcm(a, b):
    return abs(a * b) // gcd(a, b)


x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
print("GCD:", gcd(x, y))
print("LCM:", lcm(x, y))
```
